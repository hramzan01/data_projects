```python
import csv
import os
import shutil
```

```python
# Function to duplicate a file with a new name and the same extension
def duplicate_file_with_same_extension(src_path, dest_directory, new_filename):
    try:
        # Get the extension of the source file
        _, extension = os.path.splitext(src_path)
        
        # Construct the destination path with the new filename and the same extension
        dest_path = os.path.join(dest_directory, new_filename + extension)
        
        shutil.copy2(src_path, dest_path)
    except Exception as e:
        print(f"Error copying file: {str(e)}")

# Path to the directory where your source file is located
source_directory = "C:/Users/hramzan/Desktop/preview_folder/exports/copy_reference"

# Path to the directory where you want to store duplicated files
destination_directory = "C:/Users/hramzan/Desktop/preview_folder/exports"

# Path to the CSV file containing new filenames
csv_file = "names.csv"

# Open the CSV file and read the new filenames
with open(csv_file, 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    
    # Assuming the new filenames are in the first column of the CSV
    for row in csv_reader:
        if len(row) > 0:
            new_filename = row[0]
            
            # Construct the full source path
            src_path = os.path.join(source_directory, "RSHP_Detail_Tap-01.rfa")  # Replace "original_file.ext" with the actual source filename
            
            # Check if the source file exists
            if os.path.exists(src_path):
                # Duplicate the file with the new name and the same extension
                duplicate_file_with_same_extension(src_path, destination_directory, new_filename)
                print(f"Duplicated: {new_filename}")
            else:
                print(f"Source file not found: {src_path}")

print("Batch duplication complete.")
```
